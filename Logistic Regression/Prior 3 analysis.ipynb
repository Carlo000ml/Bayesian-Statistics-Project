### Analysis of the prior for model 3

\begin{equation}  
   Y_i | \mu_i \sim Be(\mu_i)\\
   \mu_i = inv-logit(x_i^t \beta) \\
   \beta_i  = z N(-2.5 , 1) +(1-z) N(2.5,1) \\
   z\sim Beta(0.5,0.5)
\end{equation}

###### Analysis of the beta distrubution

Parameters equal to 0.5 with the aim of assiging values close to 0 or to 1.



```r
z=rbeta(10000000,0.5,0.5)
```

```r
plot(density(z))
```

###### Analysis of the beta distribution
Sample from the distribution of beta using Augmentation Trick.
Mixture of two Normal distributions with wights given by the beta distribution. The aim is to assign more mass to values of beta different from 0.

```r
beta= z* rnorm(10000000,-2.5,1)+(1-z)*rnorm(10000000,2.5,1)
```

```r
plot(density(beta))
```
